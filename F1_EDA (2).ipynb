```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
df = pd.read_csv('sample_data/FinalFeatureSet.csv')
df.head()
```

```python
plt.figure(figsize=(10,4))
sns.heatmap(df.isna(), cbar=False)
plt.title('Missing Value Heatmap')
plt.show()
df.isna().sum()
```

```python
numeric = df.select_dtypes(include=np.number).columns
df[numeric].hist(figsize=(14,10), bins=20)
plt.suptitle('Distribution of Numerical Features', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(df[numeric].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["Q1_s"], kde=True, bins=40, color="blue", label="Q1")
sns.histplot(df["Q2_s"], kde=True, bins=40, color="orange", label="Q2")
sns.histplot(df["Q3_s"], kde=True, bins=40, color="green", label="Q3")
plt.legend()
plt.title("Qualifying Time Distributions")
plt.xlabel("Qualifying Time (seconds)")
plt.show()
```

```python
sns.histplot(df["grid"], bins=20, kde=False)
plt.title("Distribution of Grid Positions")
plt.xlabel("Grid Position")
plt.show()
```

```python
sns.scatterplot(data=df, x="grid", y="target_finish_position", alpha=0.5)
plt.title("Grid Start vs Final Position")
plt.xlabel("Grid")
plt.ylabel("Finish Position")
plt.show()
```

```python
sns.histplot(df["team_reliability"], kde=True)
plt.title("Team Reliability Distribution")
plt.show()
```

```python
sns.histplot(df["driver_avg_finish_last3"], kde=True)
plt.title("Driver: Average Finish (Last 3 Races)")
plt.show()
```

```python
sns.histplot(df["track_overtaking"], kde=True)
plt.title("Track Overtaking Difficulty Score Distribution")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,5))

sns.histplot(df["expected_temp"], kde=True, ax=ax[0])
ax[0].set_title("Expected Temperature")

sns.histplot(df["expected_rain_prob"], kde=True, ax=ax[1])
ax[1].set_title("Rain Probability")

plt.show()
```
